```python
from pathlib import Path
import sys
import numpy as np
import itk

parent_dir = Path.cwd().parent
sys.path.insert(0, str(parent_dir))

import ciclope

from utils.img_io import read_image_metadata, read_image
```

```python
img_path = Path("../../00_DATA/00_test/derived/Sample1_segmented.mha")

# bone_arr_path = img_path.with_suffix(".npy")
# bone_arr = np.load(bone_arr_path)

bone, _ = read_image(img_path)
# downsample image factor DS
DS = 2
bone_ds = itk.shrink_image_filter(bone, shrink_factors=[DS, DS, DS])
bone_arr = itk.array_from_image(bone_ds)
```

```python
# Generate voxel mesh
img_metadata = read_image_metadata(img_path)
spacing = [bone_ds.GetSpacing()[0], bone_ds.GetSpacing()[1], bone_ds.GetSpacing()[2]]

mesh = ciclope.core.voxelFE.vol2ugrid(bone_arr, spacing, verbose=True)
mesh.cell_data["GV"] = [mesh.cell_data["GV"][0].astype("uint8")]
mesh.write(img_path.with_suffix(".vtu"))
```
