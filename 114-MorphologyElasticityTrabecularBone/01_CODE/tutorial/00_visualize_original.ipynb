```python
import itk
import numpy as np
import matplotlib.pyplot as plt
```

```python
impath = "Sample1.ISQ"
image = itk.imread(impath)
image_np = itk.array_from_image(image)
```

```python
plt.imshow(image_np[np.shape(image_np)[0] // 2, :, :], cmap="gray")
# no labels
plt.axis("off")
plt.show()
```
